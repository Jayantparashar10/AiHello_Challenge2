```python
import pandas as pd
import numpy as np
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

def extract_features(text, tokenizer, model):
    """Extract BERT features from text"""
    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True, padding='max_length')
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

def calculate_requirement_score(employer, candidate):
    """Calculate score based on specific requirements"""
    score = 0
    total_criteria = 0

    # Salary match
    if employer.get('salary_range') and candidate.get('salary_desired'):
        total_criteria += 1
        if employer['salary_range'][0] <= candidate['salary_desired'] <= employer['salary_range'][1]:
            score += 1

    # Education match
    if employer.get('education') and candidate.get('education'):
        total_criteria += 1
        if candidate['education'].lower() in employer['education'].lower():
            score += 1

    # Experience match
    if employer.get('experience') is not None:
        total_criteria += 1
        candidate_exp = candidate.get('experience', 0)
        if candidate_exp >= employer['experience']:
            score += 1

    # Location match
    if employer.get('location') and candidate.get('location'):
        total_criteria += 1
        if candidate['location'].lower() in employer['location'].lower():
            score += 1

    # If no criteria were checked, return 1 to not penalize missing information
    return score / total_criteria if total_criteria > 0 else 1

def rank_candidates(employers, candidates):
    """Rank candidates for each job posting"""
    # Initialize BERT
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased')

    rankings = {}

    for employer in employers:
        print(f"\nAnalyzing candidates for: {employer['title']}")
        rankings[employer['title']] = []

        # Get employer job description features
        job_features = extract_features(employer['description'], tokenizer, model)

        for candidate in candidates:
            # Get candidate description features
            candidate_features = extract_features(candidate['description'], tokenizer, model)

            # Calculate text similarity score
            similarity_score = cosine_similarity(
                job_features.reshape(1, -1),
                candidate_features.reshape(1, -1)
            )[0][0]

            # Calculate requirements match score
            req_score = calculate_requirement_score(employer, candidate)

            # Calculate final score (weighted average)
            final_score = (similarity_score * 0.6) + (req_score * 0.4)

            rankings[employer['title']].append({
                'name': candidate['name'],
                'similarity_score': similarity_score,
                'requirement_score': req_score,
                'final_score': final_score
            })

        # Sort candidates by final score
        rankings[employer['title']] = sorted(
            rankings[employer['title']],
            key=lambda x: x['final_score'],
            reverse=True
        )

    return rankings

def print_rankings(rankings):
    """Print formatted rankings for each job"""
    for job_title, candidates in rankings.items():
        print(f"\n{'='*80}")
        print(f"Job Position: {job_title}")
        print(f"{'='*80}")
        print(f"{'Rank':<6}{'Name':<15}{'Text Match':<15}{'Req Match':<15}{'Final Score':<15}")
        print('-' * 80)

        for i, candidate in enumerate(candidates, 1):
            print(f"{i:<6}{candidate['name']:<15}{candidate['similarity_score']:.3f}{'':>7}"
                  f"{candidate['requirement_score']:.3f}{'':>7}{candidate['final_score']:.3f}{'':>7}")
        print()

# Main execution
if __name__ == "__main__":
    rankings = rank_candidates(employers, candidates)
    print_rankings(rankings)
```
